### First we must determine the architecture

I don't want to tune that many parameters therefore I'll use model with 3 conv layers, 2 dense and 1 output. The goal will be to find good values of units in a layer and what dropout to use.

```python
def exp_range(end, start = 16):
    step = 16
    while True:
        yield start
        start += step
        step *=2
        if start > end:
            break
```

```python
archs = {}
archs['con'] = exp_range(end=512)
archs['drop'] = range(4)
archs['dense'] = exp_range(end=1024)
```

```python
def prepare_model(con, drop, dense):
    model = tf.keras.Sequential([
    Conv2D(con, 3, input_shape=(28,28,1), activation = 'relu'),
    Conv2D(con, 5,activation = 'relu'),
    BatchNormalization(),
    
    Conv2D(con, 3, padding = 'same' ,activation = 'relu'),
    AveragePooling2D(pool_size=(2,2)),
    BatchNormalization(),
    
    Conv2D(con, 3, padding = 'same' ,activation = 'relu'),
    AveragePooling2D(pool_size=(2,2)),
    BatchNormalization(),
    
    Flatten(),
    Dense(dense, activation='relu'),
    Dropout(drop),
    Dense(dense, activation='relu'),
    Dropout(drop),
    Dense(10, activation='softmax'),
])
    return model
    
def find_best(archs, gen):
    con = archs['con']
    drop = archs['drop']
    dense = archs['dense']
    
    so_far_best_loss = inf
    tested_archs = []
    
    for n_con in con:
        print("New CON")
        for n_dense in dense:
            print("New DENSE")
            for n_drop in drop:
                model = prepare_model(n_con, n_drop/10, n_dense)
                model.compile(optimizer = Adam(), loss = CategoricalCrossentropy(from_logits = True), metrics = ['accuracy'])
    
                history = model.fit(gen.flow(x_train,y_train, batch_size = 16), steps_per_epoch = len(x_train)/16,
                              epochs=20, validation_data = (x_val, y_val), verbose = 0)
                current_val_loss = min(history.history['val_loss'])
                tested_archs.append({'loss':current_val_loss, 'con':n_con,'dense':n_dense,'drop':n_drop})
                if current_val_loss < so_far_best_loss:
                    so_far_best_loss = current_val_loss
                    print(f'Best val_loss {current_val_loss}')
                    print(f"And its accuracy {history.history['val_accuracy'][history.history['val_loss'].index(current_val_loss)]}")
                    with open('tested_archs.txt', 'w') as f:
                        f.write(str(tested_archs))
                
                
    return tested_archs
```

```python
gen = ImageDataGenerator(rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range = 0.1
                        )
gen.fit(x_train)
best = find_best(archs,gen)
```

### Final model

```python
con = 32
dense = 64
drop = 0.3


model = tf.keras.Sequential([
    Conv2D(con, 3, input_shape=(28,28,1), activation = 'relu'),
    Conv2D(con, 5,activation = 'relu'),
    BatchNormalization(),
    
    Conv2D(con, 3, padding = 'same' ,activation = 'relu'),
    AveragePooling2D(pool_size=(2,2)),
    BatchNormalization(),
    
    Conv2D(con, 3, padding = 'same' ,activation = 'relu'),
    AveragePooling2D(pool_size=(2,2)),
    BatchNormalization(),
    
    Flatten(),
    Dense(dense, activation='relu'),
    Dropout(drop),
    Dense(dense, activation='relu'),
    Dropout(drop),
    Dense(10, activation='softmax'),
])
```

```python
model.summary()
```

```python
model.compile(optimizer = Adam(amsgrad=True), loss = CategoricalCrossentropy(from_logits = True), metrics = ['accuracy'])
```

```python
train = pd.read_csv("digit-recognizer/train.csv")
test = pd.read_csv("digit-recognizer/test.csv")
```

```python
y_train = train["label"]
x_train = train.drop(labels = ["label"],axis = 1) 


y_train.value_counts()

x_train = x_train / 255.0
test = test / 255.0

y_train = to_categorical(y_train, num_classes = 10)

x_train = np.array(x_train).reshape(42000,28,28,1)
```

```python
test = np.array(test).reshape(28000,28,28,1)
```

```python
del train 
```

```python
def get_new_model():
    con = 32
    dense = 64
    drop = 0.3


    model = tf.keras.Sequential([
        Conv2D(con, 3, input_shape=(28,28,1), activation = 'relu'),
        Conv2D(con, 5,activation = 'relu'),
        BatchNormalization(),

        Conv2D(con, 3, padding = 'same' ,activation = 'relu'),
        AveragePooling2D(pool_size=(2,2)),
        BatchNormalization(),

        Conv2D(con, 3, padding = 'same' ,activation = 'relu'),
        AveragePooling2D(pool_size=(2,2)),
        BatchNormalization(),

        Flatten(),
        Dense(dense, activation='relu'),
        Dropout(drop),
        Dense(dense, activation='relu'),
        Dropout(drop),
        Dense(10, activation='softmax'),
    ])
    model.compile(optimizer = Adam(amsgrad=True), loss = CategoricalCrossentropy(from_logits = True), metrics = ['accuracy'])
    
    return model
```

```python
gen = ImageDataGenerator(rotation_range=12,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range = 0.1
                        )
gen.fit(x_train)
```

```python
models = []
```

```python
n_models = 70
epochs = 40

for i in range(n_models):
    model = get_new_model()
    model.fit(gen.flow(x_train,y_train, batch_size = 16), steps_per_epoch = len(x_train)/16,
        epochs=epochs, verbose = 1)
    models.append(model)
    model.save(f'MNIST{i}')
```

```python
results = models[0].predict(test)
for m in models[1:]:
    results += m.predict(test)

results = np.argmax(results,axis = 1)

results = pd.Series(results,name="Label")

submission = pd.concat([pd.Series(range(1,28001),name = "ImageId"),results],axis = 1)

submission.to_csv("out.csv",index=False)
```
