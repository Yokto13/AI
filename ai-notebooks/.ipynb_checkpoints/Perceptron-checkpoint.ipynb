```python
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
class Perceptron:
    def __init__(self):
        self.data = None
        self.w = None
        
    def init_w(self, sz):
        self.w = [0] * (sz + 1)

    def add(self, u, v):
        assert len(u) == len(v)
        return [a+b for a,b in zip(u, v)]
    
    def train(self, data, patience=100):
        self.data = data
        self.init_w(len(data[0]))
        all_correct = False
        n_iters = 0
        while not all_correct and n_iters < patience:
            all_correct = True
            for (x,t) in data:
                # We are hiding bias in input examples.
                # Thus we need to care only about updating weights.
                y = self.dot(x + [1],self.w)
                # wrong sample:
                if t*y <= 0:
                    self.w = self.add(self.w, [t*e for e in x] + [t])
                    all_correct = False
            n_iters += 1
        # if we ran out of patience...
        if not n_iters < patience:
            print("Warning")
            print("Patience ran out.")
            print("Data might not be linearly separable")

    def dot(self, u, v):
        assert len(u) == len(v)
        return sum([a*b for a, b in zip(u, v)])

    def predict(self, x):
        return dot(x + [1], self.w)

    def plot2d(self, data=None, w=None):
        def get_y(x1):
            # w[2] is the hidden bias
            return -(x1 * self.w[0] + self.w[2])/self.w[1]
        if data is None:
            data = self.data
        if w is None:
            w = self.w
        assert len(data[0]) == 2
        red = [x for (x,y) in data if y == 1]
        blue = [x for (x,y) in data if y == -1]
        plt.plot([e[0] for e in red], [e[1] for e in red], 'rx')
        plt.plot([e[0] for e in blue], [e[1] for e in blue], 'bo')
        left = min(
            min((x1 for (x1, x2) in red)),
            min((x1 for (x1, x2) in blue)))
        right = max(
            max((x1 for (x1, x2) in red)),
            max((x1 for (x1, x2) in blue)))
        reso = 1000
        print("left", left)
        print("right", right)
        boundary_xs = [i/reso for i in range(int(left*reso), int(right*reso))]
        boundary_ys = [get_y(x1) for x1 in boundary_xs]
        plt.plot(boundary_xs, boundary_ys)
        plt.show()


def perceptron_example(n_of_points):
    perceptron = Perceptron()
    X, Y = datasets.make_blobs(n_samples=n_of_points, centers=2, n_features=2, center_box=(0, 10))
    data = [(list(x), y if y == 1 else -1) for x,y in zip(X, Y)]
    perceptron.train(data)
    perceptron.plot2d()
```

# Examples
## Basic example

```python
perceptron_example(6)
```

## More points

```python
perceptron_example(60)
```

## Why it's bad?

```python
perceptron = Perceptron()
data = [([1, 10], 1), ([1, 1], -1), ([2, 1], -1)]
perceptron.train(data)
perceptron.plot2d()
```

There is a very little *safe space* for blue points. Obviously, it would make more sense to draw the line somewhere at the middle between red and blue points. 

But why it is so bad? The perceptron algorithm pays no attention to a so called margin, thus for some inputs the decision boundary might be so horrible. This problem is adressed by hard-margin SVMs that try to maximaze the distance to the closest points to the line. 

## Circles
Note that for data that are not linearly-separable the output makes no sense.

```python
perceptron = Perceptron()
X, Y = datasets.make_circles(n_samples=100)
data = [(list(x), y if y == 1 else -1) for x,y in zip(X, Y)]
perceptron.train(data)
perceptron.plot2d()
```
