```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def graph(x, y):
    fig = plt.figure()
    fig.set_figwidth(8)
    plt.plot(x,y, 'b')
    plt.show()
```

```python
x = np.linspace(-10, 10, 500)
```

```python
# ReLU
y = np.max((x, np.zeros(500)), axis=0)
graph(x, y)
```

```python
# sigmoid
y = 1 / (1 + np.exp(-x))
graph(x, y)
```

```python
# derivative of sigmoid
y = (1 / (1 + np.exp(-x))) * (1 - 1 / (1 + np.exp(-x)))
graph(x, y)
```

```python
# tanh
y = 2* (1 / (1 + np.exp(-x))) - 1
graph(x, y)
```

```python
# derivative of tanh
y = 1 - (2* (1 / (1 + np.exp(-x))) - 1) ** 2
graph(x, y)
```
